```python
import pandas as pd
import numpy as np
```

```python
edu_files = []
for i in range(2011, 2021):
    edu_files.append("edustat"+str(i)+".xlsx")
```

```python
edu_files
```

```python
edustat = pd.DataFrame()
edustat_final = pd.DataFrame()
for i in range(10):
    edustat = pd.read_excel(edu_files[i])
    edustat_final = pd.concat([edustat_final, edustat], ignore_index=True)
```

```python
edustat_final.head()
```

```python
edustat_final["year"] = edustat_final["조사기준일"].str.slice(start=0, stop=4)
```

```python
edustat_final["year"].value_counts
```

```python
filtered_df = edustat_final[edustat_final.시도.isin(["서울", "경남", "부산", "울산"])]
```

```python
filtered_df.to_excel("edustat_부울경.xlsx")
```

```python
filtered_df.columns
```

```python
filtered_df['졸업자중남자'] = filtered_df['졸업자_남'] / filtered_df['졸업자_계']
filtered_df['졸업자중여자'] = filtered_df['졸업자_여'] / filtered_df['졸업자_계']
filtered_df['취업자중남자'] = filtered_df['취업자_합계_남'] / filtered_df['취업자_합계_계']
filtered_df['취업자중여자'] = filtered_df['취업자_합계_여'] / filtered_df['취업자_합계_계']
filtered_df['남자취업률'] = filtered_df['취업자_합계_남'] / filtered_df['졸업자_남']
filtered_df['여자취업률'] = filtered_df['취업자_합계_여'] / filtered_df['졸업자_여']
```
